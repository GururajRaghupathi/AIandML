```python
import heapq

initial_state = [7,2,4,5,-1,6,8,3,1]
final_state= [1,2,3,4,5,6,7,8,-1]

def heuristic_distance(state,goal_state):
    diff=0
    for i in range(len(state)):
        if state[i]!= goal_state[i]:
            diff+=1
    return diff

def swap(array,i,j):
    temp=array[i]
    array[i]=array[j]
    array[j]=temp

def next_states(state):
    states=[]
    index = state.index(-1)
    if(index+3 <9):
        state1 = list(state)
        swap(state1,index,index+3)
        states.append(state1)
    if(index-3>-1):
        state2 = list(state)
        swap(state2,index,index-3)
        states.append(state2)
    if(index+1)%3!=0:
        state3 = list(state)
        swap(state3,index,index+1)
        states.append(state3)
    if(index%3!=0):
        state4 = list(state)
        swap(state4,index,index-1)
        states.append(state4)
    return states

def printPuzzle(state):
    state1 = list(state)
    index = state1.index(-1)
    state1[index]=' '
    print(state1[0],state1[1],state1[2])
    print(state1[3],state1[4],state1[5])
    print(state1[6],state1[7],state1[8])

    
class Position():
    def __init__(this,state,parent,final_state):
        this.state = state
        this.parent= parent
        this.closed= False
        this.hscore = heuristic_distance(this.state,final_state)        
        this.gscore =  1 if(not this.parent) else this.parent.gscore+1
        this.fscore = this.hscore+this.gscore
        
    def __lt__(self,other):
        return self.fscore<other.fscore
        
def solveByAStar(initial_state,final_state):
    state = initial_state
    heap = [Position(initial_state,None,final_state)]
    heapq.heapify(heap)
    visited_states= {tuple(initial_state)}
    flag=True
    while(flag):
        min_position = heapq.heappop(heap)   
        state= min_position.state
        states = next_states(state)
        for child_state in states:
            if(tuple(child_state) not in visited_states):
                visited_states.add(tuple(child_state))
                p = Position(child_state,min_position,final_state)
                heapq.heappush(heap,p)
                if(child_state==final_state):
                    flag=False
                    final_position = p
                    break
    steps=-2
    while(final_position):
        steps+=1
        printPuzzle(final_position.state)
        print()
        final_position=final_position.parent
    print("Steps taken",steps)

solveByAStar(initial_state,final_state) 
```
